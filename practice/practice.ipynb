```python
# 1. Take a number as input and print whether it is even or odd.
def num_type():
    num= int(input("enter a number: "))
    if num%2==0:
        print("number is even")
    else:
        print("number is odd")
num_type()
```

```python
# 2. Swap two variables without using a third variable
def swap_nums(num1,num2):
    print("Numbers before swap: ", num1,num2)
    num1 = num1 + num2
    num2 = num1 - num2
    num1 = num1 - num2
    print("swaped number:", num1, num2)

swap_nums(10,20)
```

```python
# 3. Convert a temperature from Celsius to Fahrenheit. 
# Formula [(°F) =  (°C) * 9/5) + 32]
def temp_conversion(c):
    f = c * (9/5) + 32 
    print(f)

result= temp_conversion(-100)
```

```python
# 4. Take a string and print it in reverse using slicing. [Interview Important]
def reverse_string(text):
    print("String:", text)
    print("Reversed String:", text[::-1])

reverse_string("abhay")
```
